## Loading Libraries and Dataset ##

```python
import pandas as pd
import numpy as np
import re

# Loading Datasets
DATA_SET1 = './IMDBDataset_d1.csv'
df1 =pd.read_csv(DATA_SET1)

DATA_SET2 = './labelled_sen_d2.csv'
df2 =pd.read_csv(DATA_SET2)

DATA_SET3 = './twitter_d3.csv'
df3 =pd.read_csv(DATA_SET3)

DATA_SET4 = './sentiment140_d4.csv'
df4 =pd.read_csv(DATA_SET4, names=['polarity', 'id', 'date', 'query', 'user', 'text'], encoding='latin-1')
```

## Dataset 1: IMDB Dataset ##

```python
df1.sample(10)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
sentiment_label = le.fit_transform(df1["sentiment"])
print(le.classes_)
df1 =df1.drop("sentiment",axis="columns")
df1["sentiment"] = sentiment_label
df1.sample(10)
# negative - 0 , positive - 1
```

## Dataset 2: Sentiment Labelled Sentences ##

```python
df2.sample(10)
```

## Dataset 3: Twitter ##

```python
df3.sample(10)
```

```python
df3 = df3.drop(columns=['2401', 'Borderlands'])
df3.dropna(inplace=True)
df3.sample(10)
```

```python
df3 = df3[df3['Positive'] != 'Irrelevant']
df3 = df3[df3['Positive'] != 'Neutral']
df3.shape
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
sentiment_label = le.fit_transform(df3["Positive"])
print(le.classes_)
df3 =df3.drop("Positive",axis="columns")
df3["Positive"] = sentiment_label
df3.rename(columns={'im getting on borderlands and i will murder you all ,': 'review', 'Positive': 'sentiment'}, inplace=True)
df3.sample(10)
# negative - 0 , positive - 1
```

## Dataset 4: Sentiment140 ##

```python
df4.sample(10)
```

```python
df4 = df4.drop(columns=['id', 'date', 'query', 'user'])
df4.rename(columns={'polarity': 'sentiment', 'text': 'review'}, inplace=True)
new_col = ['review','sentiment']
df4 = df4[new_col]
df4.sample(10)
```

```python
df4 = df4.sample(n=50000)
df4.sentiment = df4.sentiment.replace({0: 0, 4: 1})
df4.sentiment.value_counts()
```

```python
dfs = [df1,df2,df3,df4]

final_dataset = pd.concat(dfs) 

final_dataset
```

```python
print('df1:',df1.shape)
print('df2:',df2.shape)
print('df3:',df3.shape)
print('df4:',df4.shape)
print('\nFinal Dataset:',final_dataset.shape)

final_dataset.to_csv('FinalDataset.csv')
```
